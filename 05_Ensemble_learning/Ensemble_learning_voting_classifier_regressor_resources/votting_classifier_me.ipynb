```python
import pandas as pd

df = pd.read_excel("Raisin_Dataset.xlsx")
df.sample(5)
```

```python
X = df[["Area", "MajorAxisLength", "MinorAxisLength", "Eccentricity", "ConvexArea", "Extent", "Perimeter"]]
y = df["Class"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

### Model Training Using SVM: RBF Kernel

```python
from sklearn.svm import SVC

model = SVC(kernel="rbf")
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)

model.n_iter_
```

#### Model Training Using multiple model using voting_classifier

```python
from sklearn.ensemble import VotingClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

clf1  =  LogisticRegression(max_iter = 2000)
clf2  = SVC()
clf3  = DecisionTreeClassifier()

eclf  = VotingClassifier( estimators=[ ('lr', clf1), ('svc', clf2), ('dtc', clf3)]
                          , voting='hard')

eclf.fit(X_train , y_train)

y_predict  = eclf.predict(X_test)

print(classification_report(y_predict , y_test))
```

```python
from sklearn.ensemble import VotingClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

clf1  =  LogisticRegression(max_iter = 2000)
clf2  = SVC()
clf3  = DecisionTreeClassifier()

eclf  = VotingClassifier( estimators=[ ('lr', clf1), ('svc', clf2), ('dtc', clf3)]
                          , voting='soft')

eclf.fit(X_train , y_train)

y_predict  = eclf.predict(X_test)

print(classification_report(y_predict , y_test))
```
